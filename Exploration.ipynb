NOTE: Notice that in the boxplot above it is difficult to determine if there is a meaningful difference in reaction time across the 3 judgement types. So how do we know if there is a real difference? 

We rely on statistics for this. The reasoning goes like this:

Assume there is no difference between the 3 groups at all. This would mean they have identical distributions of reaction times. Given there is no real difference between the 3, what is the probability that we obtain the results that we actually recorded in our data above? 

If it is VERY likely that we obtain the data that we obtain, we could safely assume that there's no difference between the groups because our data fits well with our assumption of no difference. 

However, if it is VERY UNLIKELY that we obtain our data given the assumption that the groups do not have real differences in reaction time, then we would have to reject that assumption and conclude that THERE ARE DIFFERENCES in reaction time between the 3 judgement types. 

For more help on this, google "t-test explanation".

```r
# in this block of code we run our t-test. You don't need to understand the code our most of the output now
# except for the p-value we obtain (at the very bottom)
library(lmerTest)
# lineaer mixed effects model to test for significance of cue type on RT
# allow random intercepts for subject
lmer_mod <- lmer(event03_rating_RT ~ as.numeric(event01_block_cue_type)  + (1|src_subject_id) , data= df, REML = F)
summary(lmer_mod)
anova(lmer_mod)
```

NOTE: The final value above, Pr(>F) is our p-value. Because it is below a cutoff of 0.05, we conclude that there is a statistically significant difference between the reaction times of the 3 judgement types. 

This is important because this indicates that one of the judgement types was significantly MORE DIFFICULT than the others. We will have to take this into account in our future analyses.

It is also a finding of its own that is interesting that you could talk about in your presentation. Basically, "three common cognitive processes of interest in neuroscience are self-referential processing, likeability, and mentalizing. We wanted to determine if any of these processes are more difficult than the others. To test this, we had fMRI participants complete many trials of all three judgement types. We measured participants' reaction times while completing these trials. I then conducted a statistical test (a linear mixed effects model) to test if difficulty (measured as reaction times) were different for the 3 cognitive processes. I found that there was a statistically significant difference (p = 0.038) in reaction times. This difference in difficulty suggests that these three cognitive processes are distinct from one another at the level of the brain.
